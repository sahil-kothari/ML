```python
from apyori import apriori
```

```python
df=pd.read_csv(r'\Users\shekhar\Desktop\ML Dataset\Apriori\Order1.csv')
```

```python
df
```

```python
transactions = []
for i in range(0, 10):
    transactions.append([str(df.values[i,j]) for j in range(0,3)])
```

```python
rule_list = apriori(transactions, min_support = 0.003, min_confidence = 0.3, min_lift = 3, min_length = 2)
```

```python
results = list(rule_list)
for i in results:
    print('\n')
    print(i)
    print('**********') 
```
