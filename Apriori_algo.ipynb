```python
from apyori import apriori
```

```python
transaction=[]
for i in range (0,7500):
  transaction.append([str(df.values[i,j])for j in range(0,20)])
```

```python
rule_list=apriori(transaction,min_support=0.005,min_confidence=0.45,min_lift=3,min_length=2)
```

```python
result=list(rule_list)
for i in result:
      print('\n')
      print(i)
      print('*********************************************************************************************************************************')
```
