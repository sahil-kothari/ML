```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Pregnancies']=df['Pregnancies'].astype(float)
df['Glucose']=df['Glucose'].astype(float)
df['BloodPressure']=df['BloodPressure'].astype(float)
df['SkinThickness']=df['SkinThickness'].astype(float)
df['Insulin']=df['Insulin'].astype(float)
df['Age']=df['Age'].astype(float)
df['Outcome']=df['Outcome'].astype(float)
```

```python
df.dtypes
```

```python
df.columns
```

```python
X = df.drop(['Outcome'],axis=1)
```

```python
X
```

```python
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
y_predict=clf.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print("Confusion Matrix is : \n",metrics.confusion_matrix(y_test,y_predict))
```

```python
print(f"Accuracy score is : {metrics.accuracy_score(y_predict,y_test)}")
print(f"F1 score is : {metrics.f1_score(y_predict,y_test)}")
print(f"Precision score is : {metrics.precision_score(y_predict,y_test)}")
print(f"Recall score is : {metrics.recall_score(y_predict,y_test)}")
```
