```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
mean = df['LoanAmount'].mean()
df['LoanAmount'].fillna(value=mean, inplace=True)
```

```python
mode = df['Gender'].mode().iloc[0]
df['Gender'].fillna(value=mode, inplace=True)

mode = df['Married'].mode().iloc[0]
df['Married'].fillna(value=mode, inplace=True)

mode = df['Dependents'].mode().iloc[0]
df['Dependents'].fillna(value=mode, inplace=True)

mode = df['Self_Employed'].mode().iloc[0]
df['Self_Employed'].fillna(value=mode, inplace=True)

mode = df['Loan_Amount_Term'].mode().iloc[0]
df['Loan_Amount_Term'].fillna(value=mode, inplace=True)

mode = df['Credit_History'].mode().iloc[0]
df['Credit_History'].fillna(value=mode, inplace=True)
```

```python
df['Loan_ID'] = df['Loan_ID'].factorize()[0]
```

```python
df['Gender'] = df['Gender'].factorize()[0]
df['Married'] = df['Married'].factorize()[0]
df['Dependents'] = df['Dependents'].factorize()[0]
df['Education'] = df['Education'].factorize()[0]
df['Self_Employed'] = df['Self_Employed'].factorize()[0]
df['Property_Area'] = df['Property_Area'].factorize()[0]
df['Loan_Status'] = df['Loan_Status'].factorize()[0]
```

```python
df
```

```python
df['Loan_ID']= df['Loan_ID'].astype(float)
df['Married']= df['Married'].astype(float)
df['Gender']= df['Gender'].astype(float)
df['Dependents']= df['Dependents'].astype(float)
df['Education']= df['Education'].astype(float)
df['Self_Employed']= df['Self_Employed'].astype(float)
df['Property_Area']= df['Property_Area'].astype(float)
df['Loan_Status']= df['Loan_Status'].astype(float)
df['ApplicantIncome']= df['ApplicantIncome'].astype(float)
df['CoapplicantIncome']= df['CoapplicantIncome'].astype(float)
df['LoanAmount']= df['LoanAmount'].astype(float)
df['Loan_Amount_Term']= df['Loan_Amount_Term'].astype(float)
df['Credit_History']= df['Credit_History'].astype(float)
```

```python
df.dtypes
```

```python
X= df.drop(['Loan_Status', 'Loan_ID'], axis=1).values
y= df['Loan_Status']
```

```python
X
```

```python
# X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print('Confusion Matrix')
print(metrics.confusion_matrix(y_test, y_pred))
```

```python
print("1. Accuracy Score:", metrics.accuracy_score(y_test, y_pred)*100)
print("2. Precision Score:",metrics.precision_score(y_test, y_pred))
print("3. Recall Score:", metrics.recall_score(y_test, y_pred))
print("4. f1 Score:", metrics.f1_score(y_test, y_pred))
```
